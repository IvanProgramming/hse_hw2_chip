```python
!mkdir bowtie2_res
```

```python
%%time
!bowtie2  -p 2 \
          --threads 16 \
          -x chromosome_index \
          -U ENCFF000AOG.fastq \
          -S bowtie2_res/res_AOG.sam
```

```python
%%time
!bowtie2  -p 2 \
          --threads 16 \
          -x chromosome_index \
          -U ENCFF000APV.fastq \
          -S bowtie2_res/res_APV.sam
```

```python
%%time
!bowtie2  -p 2 \
          --threads 16 \
          -x chromosome_index \
          -U ENCFF000APW.fastq \
          -S bowtie2_res/res_APW.sam
```

```python
!samtools view -h -S -b -o bowtie2_res/res_APW.bam bowtie2_res/res_APW.sam
```

```python
!samtools view -h -S -b -o bowtie2_res/res_AOG.bam bowtie2_res/res_AOG.sam
!samtools view -h -S -b -o bowtie2_res/res_APV.bam bowtie2_res/res_APV.sam
```

## Фильтруем чтения

```python
!wget https://github.com/Boyle-Lab/Blacklist/raw/refs/heads/master/lists/hg38-blacklist.v2.bed.gz
```

```python
!gunzip hg38-blacklist.v2.bed.gz
```

```python
!bedtools intersect -abam bowtie2_res/res_AOG.bam -b hg38-blacklist.v2.bed > bowtie2_res/res_AOG_filtered.bam
```

```python
!bedtools intersect -abam bowtie2_res/res_APV.bam -b hg38-blacklist.v2.bed -v > bowtie2_res/res_APV_filterd.bam
```

```python
!bedtools intersect -abam bowtie2_res/res_APW.bam -b hg38-blacklist.v2.bed -v > bowtie2_res/res_APW_filterd.bam
```

## Удаляем PCR реплики

```python
# Сортировка BAM-файлов
!picard SortSam I=bowtie2_res/res_AOG_filtered.bam O=bowtie2_res/res_AOG_sorted.bam SORT_ORDER=coordinate
!picard SortSam I=bowtie2_res/res_APV_filterd.bam O=bowtie2_res/res_APV_sorted.bam SORT_ORDER=coordinate
!picard SortSam I=bowtie2_res/res_APW_filterd.bam O=bowtie2_res/res_APW_sorted.bam SORT_ORDER=coordinate

# Удаление PCR-дубликатов
!picard MarkDuplicates I=bowtie2_res/res_AOG_sorted.bam O=bowtie2_res/res_AOG_dedup.bam M=bowtie2_res/res_AOG_metrics.txt REMOVE_DUPLICATES=true
!picard MarkDuplicates I=bowtie2_res/res_APV_sorted.bam O=bowtie2_res/res_APV_dedup.bam M=bowtie2_res/res_APV_metrics.txt REMOVE_DUPLICATES=true
!picard MarkDuplicates I=bowtie2_res/res_APW_sorted.bam O=bowtie2_res/res_APW_dedup.bam M=bowtie2_res/res_APW_metrics.txt REMOVE_DUPLICATES=true

# Индексация BAM-файлов после удаления дубликатов
!samtools index bowtie2_res/res_AOG_dedup.bam
!samtools index bowtie2_res/res_APV_dedup.bam
!samtools index bowtie2_res/res_APW_dedup.bam
```

```python
!macs2 callpeak --broad \
    -t bowtie2_res/res_AOG_filtered.bam bowtie2_res/res_APV_filterd.bam \
    -c bowtie2_res/res_APW_filterd.bam \
    -f BAM \
    --outdir macs2
```
